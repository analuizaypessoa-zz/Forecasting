```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("AirQualityUCI.csv", sep = ";")  ## UCL demand forecasting
df["Date"] = pd.to_datetime(df["Date"], format = '%d/%m/%Y')
```

```python
## boxplot
sns.set(style="whitegrid")
ax = sns.boxplot(x=df['PT08.S1(CO)'])
```

```python
#### plot
fig = px.line(df, x= "Date", y='PT08.S1(CO)')
fig.update_layout(
    plot_bgcolor='rgb(247, 247, 247)',
    autosize=False,
    width=800,
    height=400,
    title={
        'text': 'PT08.S1(CO)' ,
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'}
    )
fig.show()

#data = df['Enrollments'].values
```

```python
#### making a time series
ts = df.groupby("Date")['PT08.S1(CO)'].sum().rename("CO")
ts.head()
```

```python
#### droptime series outliers
#### detect outliers and treat them as NULL
T = ts
for i in np.arange(1,len(ts.index)):
  std_bef = ts[:i-1].std()
  #mean_bef =  ts[target][:i-1].mean()
  if ( ts[i] < ts[i-1] - 2*std_bef or ts[i] >  ts[i-1]+ 2*std_bef):
    T[i] = ts[i-1]
  else:
    T[i] = ts[i]
```

```python
T.name = 'target'
ts = pd.concat([ts, T], axis=1)
ts.head()
```

```python
ts.columns = ["CO_prev","CO"]
```

```python
#### plot
fig = px.line(ts, x= ts.index, y='CO')
fig.update_layout(
    plot_bgcolor='rgb(247, 247, 247)',
    autosize=False,
    width=800,
    height=400,
    title={
        'text': 'PT08.S1(CO)' ,
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'}
    )
fig.show()

#data = df['Enrollments'].values
```

#### Trainning the model

```python
from pyFTS.partitioners import Grid

fs = Grid.GridPartitioner(data=ts["CO"],npart=10)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=[15,5])

fs.plot(ax)
```
